### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
fig = plt.figure()
x = np.linspace(0, 10, 100)

# Creo una figura de 2x1 y en el axe 1 añado lo siguiente
plt.subplot(1, 2, 1) # filas, columnas, índice
plt.plot(x, np.sin(x))
plt.title("Seno") # Título

# En el axe 2 añado lo siguiente
plt.subplot(1, 2, 2) # filas, columnas, índice
plt.plot(x, np.cos(x))
plt.title("Coseno"); # Título
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
fig, ax = plt.subplots(3, 1)

ax[0].plot(x, np.sin(x), color = "red")
ax[1].plot(x, np.cos(x), color = "green")
ax[2].plot(x, np.tan(x), color = "blue");
ax[2].set_ylim(-40, 20);
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python
fig, ax = plt.subplots(2, 1)
plt.subplots_adjust(hspace = 0.5)

ax[0].plot(x, np.sinh(x), color = "red")
ax[0].set_title("Seno Hiperbólico")
ax[0].set_xlabel("x")
ax[0].set_ylabel("sinh(x)")

ax[1].plot(x, np.cosh(x), color = "blue")
ax[1].set_title("Coseno Hiperbólico")
ax[1].set_xlabel("x")
ax[1].set_ylabel("cosh(x)");
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
fig, ax = plt.subplots(2, 2)
plt.subplots_adjust(hspace = 0.5, wspace = 0.5)

ax[0, 0].plot(x, np.sin(x))
ax[0, 1].plot(x, np.cos(x))
ax[1, 0].plot(x, np.tan(x))
ax[1, 1].plot(x, np.tanh(x));
```
