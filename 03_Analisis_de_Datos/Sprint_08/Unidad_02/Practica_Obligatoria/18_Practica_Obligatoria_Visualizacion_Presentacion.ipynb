### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersin que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def plot_scatter(df, num_col1, num_col2, cat_col = None, size = 50, scale = 1, legend = True, alpha = 0.5, palette = "viridis"):

    '''
    Pinta diagramas de dispersin dadas 2 columnas numricas de un DataFrame.
    Parmetros:
        - df: DataFrame
        - num_col1: Nombre de la primera columna numrica (eje x)
        - num_col2: Nombre de la segunda columna numrica (eje y)
        - cat_col: Nombre de la columna categrica para los colores (opcional)
        - size: Nombre de la columna para el tamao de los puntos o un valor constante
        - scale: Escala para el tamao de los puntos si se usa una columna para el tamao
        - legend: Booleano para mostrar la leyenda de colores y tamaos
        - alpha: Nivel de transparencia (de 0 a 1)
        - color_palette: Paleta de colores con la que quiero elaborar el diagrama
    '''

    # Determinamos el tamao de los puntos  
    if isinstance(size, str):     # Si 'size' es una cadena, se interpreta como el nombre de una columna del DataFrame
        sizes = df[size] * scale  # Los valores de esta columna se multiplican por el scale para obtener los tamaos de los puntos
    else:
        sizes = size  # Si 'size' es un nmero, todos los puntos tendrn ese tamao

    # Creamos la figura
    plt.figure(figsize = (10, 6))

    # Dibujamos el scatterplot
    if cat_col: # Si se proporciona 'cat_col', se usa como el parmetro 'hue' en 'sns.scatterplot' para colorear los puntos segn las categoras
        scatter = sns.scatterplot(data = df, x = num_col1, y = num_col2, hue = cat_col, size = sizes, sizes = (1, 250), legend = legend, alpha = alpha, palette = palette)
    else: # Si no se proporciona 'cat_col', se omite el parmetro 'hue'
        scatter = sns.scatterplot(data = df, x = num_col1, y = num_col2, size = sizes, sizes = (1, 250), legend = legend, alpha = alpha)

    # Leyenda
    if legend:
        if cat_col:
            plt.legend(loc = 'best')
        else: 
            scatter.figure.colorbar(scatter.collections[0]) # Si no se usa 'cat_col', se muestra un colorbar para los tamaos de los puntos.
    
    plt.title(f"Diagrama de dispersin de {num_col1} vs {num_col2}")
    plt.xlabel(num_col1)
    plt.ylabel(num_col2)
    plt.show()
```

```python
# Cargar el dataset del Titanic
df_titanic = pd.read_csv("./data/titanic.csv")
df_titanic.head(5)
```

#### Grfico tasa de mortalidad

```python
# Crear un grfico de barras en frecuencias relativas
frecuencias_relativas = df_titanic["alive"].value_counts(normalize = True)*100

plt.figure(figsize = (10, 6))
sns.barplot(x = frecuencias_relativas.index, y = frecuencias_relativas.values, hue = frecuencias_relativas, palette = 'coolwarm', legend = False)

# Aadir ttulo y etiquetas
plt.title('Tasa de Mortalidad en el Titanic')
plt.xlabel('Sobrevivi')
plt.ylabel('Nmero de Pasajeros')

# Mostrar el grfico
plt.show()
```

#### Grfico 'mujeres y nios primero'

```python
# Crear un grfico de barras apilado en frecuencias relativas
plt.figure(figsize = (10, 6))
sns.countplot(data = df_titanic, x = 'who', hue = 'alive', palette = 'coolwarm', saturation = 1)

# Aadir ttulo y etiquetas
plt.title('Mortalidad en el Titanic segn Sexo y Edad')
plt.xlabel('Quin (Nio, Hombre, Mujer)')
plt.ylabel('Nmero de Pasajeros')
plt.legend(title = 'Sobrevivi')

# Mostrar el grfico
plt.show()
```

#### Grfico segn la clase

```python
# Crear un grfico de barras apilado en frecuencias relativas
plt.figure(figsize = (10, 6))
sns.countplot(data = df_titanic, x = 'class', hue = 'alive', palette = 'coolwarm', saturation = 1)

# Aadir ttulo y etiquetas
plt.title('Mortalidad en el Titanic segn la clase')
plt.xlabel('Clase')
plt.ylabel('Nmero de Pasajeros')
plt.legend(title = 'Sobrevivi')

# Mostrar el grfico
plt.show()
```

#### Grfico supervivencia mujeres y nios

```python
# Filtra los datos para mujeres y nios
filtered_titanic = df_titanic[df_titanic['who'].isin(['woman', 'child'])]

# Calcula las tasas de supervivencia
survival_rates = filtered_titanic.groupby('who')['alive'].value_counts(normalize=True).unstack() * 100
survival_rates = survival_rates.reset_index()
survival_rates.columns = ['who', 'Died', 'Survived']

# Crear el grfico de barras
plt.figure(figsize = (10, 6))
survival_plot = sns.barplot(data = survival_rates, x = 'who', y = 'Survived', hue = 'Survived', palette = 'coolwarm', legend = False)

# Aade ttulo y etiquetas
plt.title('Tasas de Supervivencia en el Titanic entre Mujeres y Nios')
plt.xlabel('Grupo')
plt.ylabel('Tasa de Supervivencia (%)')

# Aade etiquetas con los valores de porcentaje encima de las barras
for index, row in survival_rates.iterrows():
    survival_plot.text(index, row.Survived + 1, f'{row.Survived:.1f}%', color = 'black', ha = "center")
```

#### Grfico supervivencia de mujeres y nios segn la clase

```python
# Crea el grfico
plt.figure(figsize=(12, 8))
sns.catplot(data = filtered_titanic, kind = 'bar', x = 'who', hue = 'class', y = 'survived', palette = 'coolwarm', errorbar = None)
plt.title('Tasas de supervivencia de mujeres y nios segn la clase del pasaje')
plt.xlabel('Grupo')
plt.ylabel('Tasa de Supervivencia (%)')

# Muestra el grfico
plt.show()
```

#### Grfico supervivencia segn puerto de embarque y clase

```python
# Crear un grfico de barras
plt.figure(figsize = (12, 8))
sns.catplot(data = df_titanic.sort_values("class"), kind = 'bar', x = 'embark_town', hue = 'class', y = 'survived', palette = 'coolwarm', errorbar = None)

# Aadir ttulo y etiquetas
plt.title('Tasas de Supervivencia en el Titanic segn Puerto de Embarque y Clase')
plt.xlabel('Puerto de Embarque')
plt.ylabel('Tasa de Supervivencia')

# Mostrar el grfico
plt.show()
```
