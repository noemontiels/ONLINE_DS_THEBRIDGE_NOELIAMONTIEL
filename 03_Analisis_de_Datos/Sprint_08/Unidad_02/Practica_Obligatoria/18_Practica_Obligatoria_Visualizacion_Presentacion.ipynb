### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersión que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesión en vivo)

```python
def plot_scatter(df, num_col1, num_col2, cat_col = None, size = 50, scale = 1, legend = True, alpha = 0.5, palette = "viridis"):

    '''
    Pinta diagramas de dispersión dadas 2 columnas numéricas de un DataFrame.
    Parámetros:
        - df: DataFrame
        - num_col1: Nombre de la primera columna numérica (eje x)
        - num_col2: Nombre de la segunda columna numérica (eje y)
        - cat_col: Nombre de la columna categórica para los colores (opcional)
        - size: Nombre de la columna para el tamaño de los puntos o un valor constante
        - scale: Escala para el tamaño de los puntos si se usa una columna para el tamaño
        - legend: Booleano para mostrar la leyenda de colores y tamaños
        - alpha: Nivel de transparencia (de 0 a 1)
        - color_palette: Paleta de colores con la que quiero elaborar el diagrama
    '''

    # Determinamos el tamaño de los puntos  
    if isinstance(size, str):     # Si 'size' es una cadena, se interpreta como el nombre de una columna del DataFrame
        sizes = df[size] * scale  # Los valores de esta columna se multiplican por el scale para obtener los tamaños de los puntos
    else:
        sizes = size  # Si 'size' es un número, todos los puntos tendrán ese tamaño

    # Creamos la figura
    plt.figure(figsize = (10, 6))

    # Dibujamos el scatterplot
    if cat_col: # Si se proporciona 'cat_col', se usa como el parámetro 'hue' en 'sns.scatterplot' para colorear los puntos según las categorías
        scatter = sns.scatterplot(data = df, x = num_col1, y = num_col2, hue = cat_col, size = sizes, sizes = (1, 250), legend = legend, alpha = alpha, palette = palette)
    else: # Si no se proporciona 'cat_col', se omite el parámetro 'hue'
        scatter = sns.scatterplot(data = df, x = num_col1, y = num_col2, size = sizes, sizes = (1, 250), legend = legend, alpha = alpha)

    # Leyenda
    if legend:
        if cat_col:
            plt.legend(loc = 'best')
        else: 
            scatter.figure.colorbar(scatter.collections[0]) # Si no se usa 'cat_col', se muestra un colorbar para los tamaños de los puntos.
    
    plt.title(f"Diagrama de dispersión de {num_col1} vs {num_col2}")
    plt.xlabel(num_col1)
    plt.ylabel(num_col2)
    plt.show()
```

```python
# Cargar el dataset del Titanic
df_titanic = pd.read_csv("./data/titanic.csv")
df_titanic.head(5)
```

#### Gráfico tasa de mortalidad

```python
# Crear un gráfico de barras en frecuencias relativas
frecuencias_relativas = df_titanic["alive"].value_counts(normalize = True)*100

plt.figure(figsize = (10, 6))
sns.barplot(x = frecuencias_relativas.index, y = frecuencias_relativas.values, hue = frecuencias_relativas, palette = 'coolwarm', legend = False)

# Añadir título y etiquetas
plt.title('Tasa de Mortalidad en el Titanic')
plt.xlabel('Sobrevivió')
plt.ylabel('Número de Pasajeros')

# Mostrar el gráfico
plt.show()
```

#### Gráfico 'mujeres y niños primero'

```python
# Crear un gráfico de barras apilado en frecuencias relativas
plt.figure(figsize = (10, 6))
sns.countplot(data = df_titanic, x = 'who', hue = 'alive', palette = 'coolwarm', saturation = 1)

# Añadir título y etiquetas
plt.title('Mortalidad en el Titanic según Sexo y Edad')
plt.xlabel('Quién (Niño, Hombre, Mujer)')
plt.ylabel('Número de Pasajeros')
plt.legend(title = 'Sobrevivió')

# Mostrar el gráfico
plt.show()
```

#### Gráfico según la clase

```python
# Crear un gráfico de barras apilado en frecuencias relativas
plt.figure(figsize = (10, 6))
sns.countplot(data = df_titanic, x = 'class', hue = 'alive', palette = 'coolwarm', saturation = 1)

# Añadir título y etiquetas
plt.title('Mortalidad en el Titanic según la clase')
plt.xlabel('Clase')
plt.ylabel('Número de Pasajeros')
plt.legend(title = 'Sobrevivió')

# Mostrar el gráfico
plt.show()
```

#### Gráfico supervivencia mujeres y niños

```python
# Filtra los datos para mujeres y niños
filtered_titanic = df_titanic[df_titanic['who'].isin(['woman', 'child'])]

# Calcula las tasas de supervivencia
survival_rates = filtered_titanic.groupby('who')['alive'].value_counts(normalize=True).unstack() * 100
survival_rates = survival_rates.reset_index()
survival_rates.columns = ['who', 'Died', 'Survived']

# Crear el gráfico de barras
plt.figure(figsize = (10, 6))
survival_plot = sns.barplot(data = survival_rates, x = 'who', y = 'Survived', hue = 'Survived', palette = 'coolwarm', legend = False)

# Añade título y etiquetas
plt.title('Tasas de Supervivencia en el Titanic entre Mujeres y Niños')
plt.xlabel('Grupo')
plt.ylabel('Tasa de Supervivencia (%)')

# Añade etiquetas con los valores de porcentaje encima de las barras
for index, row in survival_rates.iterrows():
    survival_plot.text(index, row.Survived + 1, f'{row.Survived:.1f}%', color = 'black', ha = "center")
```

#### Gráfico supervivencia de mujeres y niños según la clase

```python
# Crea el gráfico
plt.figure(figsize=(12, 8))
sns.catplot(data = filtered_titanic, kind = 'bar', x = 'who', hue = 'class', y = 'survived', palette = 'coolwarm', errorbar = None)
plt.title('Tasas de supervivencia de mujeres y niños según la clase del pasaje')
plt.xlabel('Grupo')
plt.ylabel('Tasa de Supervivencia (%)')

# Muestra el gráfico
plt.show()
```

#### Gráfico supervivencia según puerto de embarque y clase

```python
# Crear un gráfico de barras
plt.figure(figsize = (12, 8))
sns.catplot(data = df_titanic.sort_values("class"), kind = 'bar', x = 'embark_town', hue = 'class', y = 'survived', palette = 'coolwarm', errorbar = None)

# Añadir título y etiquetas
plt.title('Tasas de Supervivencia en el Titanic según Puerto de Embarque y Clase')
plt.xlabel('Puerto de Embarque')
plt.ylabel('Tasa de Supervivencia')

# Mostrar el gráfico
plt.show()
```
